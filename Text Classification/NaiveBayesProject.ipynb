```python
import numpy as np
import os
from sklearn import model_selection as ms
import numpy as np
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import classification_report
```

```python
def convert_list_labels_tonparray(labels,c):
    dict={}
    a = np.zeros(shape=(len(labels),))
    for x in range(len(labels)):
        if labels[x] in dict:
                    a[x]=dict[labels[x]]
        else:
                    dict[labels[x]]=c
                    a[x]=dict[labels[x]]
                    c+=1
    return a
```

```python
dictionary={}
#function to change the text into array of words
def getWords(data):
    arr=[]
    for word in data.split():
        word=word.replace('\n','')
        word=word.replace('\r','')
        word=word.replace('\t','')
        word=word.replace(',','')
        word=word.replace('(','')
        word=word.replace(':','')
        word=word.replace(').','')
        word=word.replace(')','')
        word=word.replace('.','')
        word=word.replace('"','')    
        word=word.lower()
        if( word in stopword):
            continue
        else:
            arr.append(word)
            if(word in dictionary):
                dictionary[word]+=1
            else:
                dictionary[word]=1
    return arr
```

```python
stopword=[]
#creating a stop word array
s_words=open('stop.txt','r')
s_words_data=s_words.read()
for x in s_words_data.split('\n'):
    stopword.append(x)
```

```python
#defining the path of data
path = '20_newsgroups/'
data=[]
labels=[]
#going through each directory and appending it to data and labels
for itemName in os.listdir(path):
    itemName = os.path.join(path, itemName)
    print(itemName)
    for myPath in os.listdir(itemName):
        myPath=os.path.join(itemName,myPath)
        file_readed = open(myPath)
        arr=getWords(file_readed.read())
        data.append(arr)
        labels.append(itemName)
```

```python
label=convert_list_labels_tonparray(labels,0)
```

Multinomial Naive Bayes Implementation

```python
def fit(x_train,y_train):
    count={}
    all_classes=set(y_train)
    
    count["data"]=len(y_train)
    for current_class in all_classes:
        count[current_class]={}
        
        all_features=x_train.shape[1]
        current_rows=y_train==current_class
        current_x_train=x_train[current_rows]
        
        current_y_train=y_train[current_rows]
        count[current_class]["total"]=len(current_y_train)

        for current_feature in range(all_features):
            count[current_class][current_feature]=(current_x_train[:,current_feature]).sum()
    return count
```

```python
count=fit(X_train,Y_train)
print("Model Trained")
```

```python
def predict(dictionary,x_test):
    y_pred=[]
    i=1
    len_x_test=len(x_test)
    for x in x_test:
        #Print the current document under prediction
        print("document ",i," of ",len_x_test)
        i+=1
        #prediction of single document out of given len_x_test documents
        x_class=predictsinglepoint(dictionary,x)
        #append the result obtained in y_pred
        y_pred.append(x_class)
    return y_pred 
```

```python
def predictsinglepoint(dictionary,x):
    #store all the classes
    classes=dictionary.keys()
    #  variable for storing best best probabiliy and best class respectively
    best_p=-1000
    best_class=-1
    first_run=True
    for current_class in classes:
        if(current_class=="data"):
            continue
        p_current_class=probability(dictionary,x,current_class)
        if (first_run or (p_current_class > best_p)):
            best_p=p_current_class
            best_class=current_class
        first_run=False
    return best_class
```

```python
# calculation of probability for the current _class
def probability(dictionary,x,current_class):

    output=np.log(dictionary[current_class]["total"])-np.log(dictionary["data"])
    num_features=2000
    
    for j in range(1,num_features+1):
        xj=x[j-1]
        if xj!=0:
            count_current_class_with_word_xj=dictionary[current_class][j-1]+1
            count_current_class=dictionary[current_class]["total"]+2000
            current_xj_prob=np.log(count_current_class_with_word_xj)-np.log(count_current_class)
            output=output+current_xj_prob
    return output
```
